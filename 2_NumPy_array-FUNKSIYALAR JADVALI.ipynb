```python
# 2 va undan ko'p olchamli array yaratish.
data3 = [[1, 3, 4, 5], [4, 3, 8, 10], [3, 23, 53, 56]] # list ichida list list orasida (,)
arr3  = np.array(data3)
print(arr3)
```

```python
'''shape, size'''
#shape -> qator va ustunlar soni (q, u)
#size -> arrayning elementlar soni
arr3.shape # bunda 3ta qator va 4tadan son kiritilganligi uchun (3, 4) chiqaradi
```

```python
arr3.size # arr3 ni ichida 12 elemet bor
```

```python
'''zeros, ones'''
#zeros massiv elementlar barchasi 0ga teng massiv yaratadi
#ones massv elementlar barchasi 1ga teng massiv yaratadi

arr4 = np.zeros(shape = (3, 4)) #np.zeros((3, 4)) deb yozsaham bo'ladi.
arr4
```

```python
arr5 = np.ones((3, 4)) #ones ham shunaqa funksiya
arr5
```

```python
'''arange'''
# range bilan birxil ammo arrayda arange funksiyasi ihlatiladi.

arr6 = np.arange(2, 20, 4) #2dan 20gacha 4qadam tashab chiqaradi
arr6
```

```python
arr7 = np.arange(20) # 0dan 19gacha chiqaradi.
arr7
```

```python
'''rand, randint, randn'''
# rand
arr8 = np.random.rand(3, 3) #shape 3,3 o'lchamli 0~1gacha taxminiy qiymat
arr8
```

```python
# randint 
# 10 gacha bo'lgan 3,3 o'lchamli integer sonlar
arr9 = np.random.randint(10, size = (3, 3)) # shape 3,3
arr9
```

```python
# randn
arr10 = np.random.randn(3, 3) # shape 3,3 o'lchamli --> elementlar (Normal distansiali)
arr10
```
