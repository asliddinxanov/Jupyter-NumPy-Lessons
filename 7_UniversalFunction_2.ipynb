- Binary - Ikkita argument qabul qiluvchi funksiya

```python
import numpy as np
```

```python
# Ikkita array yaratib olamiz
arr1 = np.random.randn(5)
arr2 = np.random.randn(5)
```

```python
print(arr1)
print(arr2)
```

### `.add` 

```python
np.add(arr1, arr2) # ikkita arrayni bir-biriga qo'shib beradi
```

```python
arr1 + arr2 # bu ham yuqoridagini koddiy ko'rinishi
```

### `.multiply`- ikkita arrayni mos elementini ko'paytiradi 

```python
print(arr1)
print(arr2)
```

```python
np.multiply(arr1, arr2) # mos ustunlarni bir-biriga ko'paytiradi
```

### `.maximum` - ikkita ustunni taqqoslab kattasini qaytaradi

```python
print(arr1)
print(arr2)
```

```python
np.maximum(arr1, arr2) # katta ustunni qaytaradi
```
