### `sqrt`, `square`, `exp`, `log`, `modf`, `sign`, `isnan` functions

```python
import numpy as np
```

```python
arr = np.arange(10) # array
```

```python
arr
```

```python
np.sqrt(arr) # har bir elementni ildizi
```

```python
np.square(arr) # har bir elementni kvadrati
```

```python
np.exp(arr) # har bir elementni Exponentali
```

```python
arr
```

```python
np.log(arr[1:]) # logorifm
```

```python
arr1 = np.random.randn(5)
arr1
```

```python
qoldiq, butun = np.modf(arr1) # burun son va o'nli qismini ajratadi
```

```python
qoldiq # yuqoridagi arrayni o'nli qismi
```

```python
butun # yuqoridagi arrayni butun qismi
```

```python
np.sign(arr1) # massivdagi qiymatlarni musbat qimi (1) manfiy qismini (-1) chiqaradi nol (0)
```

```python
np.isnan(arr1) # NaN bo'lsa True, aks xolda False chiqaradi
```

```python
arr1[1] = np.NaN
```
