### `reshape` metodi 
`(.reshape)`

```python
import numpy as np
```

```python
a = np.arange(6).reshape(3,2) # 0~5gacha bo`lgan (3,2) li massiv
a
```

`Transpose`

```python
a.T # (.T)da massivni Trasnposr ga aylantiradi
```

# Swaping axes
`(.swapaxes)`

```python
a.swapaxes(1,0) # a ning Transpose ko`rinishi beradi
```

### `Tranpose`ning ishlatilinishiga bir misol $A A^T$
 - `(.dot)` bu bir arrayni ikinchi arrayga ko'paytirish

```python
b = np.dot(a, a.T)
b
```
